```python
from kafka import KafkaProducer
import pandas as pd
import numpy as np
import uuid
import hashlib
import json
```

```python
data = pd.read_csv("data/tweets_soft.csv", header = None, names = ["timestamp", "request", "nickname", "message"])
data["UID"] = [uuid.uuid4().hex for _ in range(data.shape[0])]
dict_date = data.to_dict("records")
data['timestamp'] = pd.to_datetime(data['timestamp'])
data['timestamp'] = data.timestamp.values.astype(np.int64)
data.sort_values(by ='timestamp')
dict_date = data.to_dict("records")

producer = KafkaProducer(bootstrap_servers='localhost:9092')
for element in dict_date:
    producer.send('twitter', str(json.dumps(element)))
```
